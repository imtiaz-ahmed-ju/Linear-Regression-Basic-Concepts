```
df['HSC_b']= ((df['PSC_b'] + df['JSC_b'] + df['SSC_b']) / 3).round(2)
```

```
df['HSC_b']= ((df['PSC_b'] + df['JSC_b'] + df['SSC_b']) / 3).round(2)
conditions = [
    (df['HSC_b'] > 4.5), (df['HSC_b'] > 4) & (df['HSC_b']<=4.5),
    (df['HSC_b'] > 3.5) & (df['HSC_b'] <= 4 ),
     (df['HSC_b'] > 3) & (df['HSC_b'] <= 3.5 ),
     (df['HSC_b'] > 2.5) & (df['HSC_b'] <= 3 ),
     (df['HSC_b'] == 2.5)
    ]
choices = ['5', '4.5', '4', '3.5', '3', '2.5']
df['HSC_b'] = np.select(conditions, choices, default='0')
print(df)
```

```
df['HSC_e']= ((df['PSC_e'] + df['JSC_e'] + df['SSC_e']) / 3).round(2)
conditions = [
    (df['HSC_e'] > 4.5), (df['HSC_e'] > 4) & (df['HSC_e']<=4.5),
    (df['HSC_e'] > 3.5) & (df['HSC_e'] <= 4 ),
     (df['HSC_e'] > 3) & (df['HSC_e'] <= 3.5 ),
     (df['HSC_e'] > 2.5) & (df['HSC_e'] <= 3 ),
     (df['HSC_e'] == 2.5)
    ]
choices = ['5', '4.5', '4', '3.5', '3', '2.5']
df['HSC_e'] = np.select(conditions, choices, default='0')
print(df)
```

```
df['HSC_m']= ((df['PSC_m'] + df['JSC_m'] + df['SSC_m']) / 3).round(2)
conditions = [
    (df['HSC_m'] > 4.5), (df['HSC_m'] > 4) & (df['HSC_m']<=4.5),
    (df['HSC_m'] > 3.5) & (df['HSC_m'] <= 4 ),
     (df['HSC_m'] > 3) & (df['HSC_m'] <= 3.5 ),
     (df['HSC_m'] > 2.5) & (df['HSC_m'] <= 3 ),
     (df['HSC_m'] == 2.5)
    ]
choices = ['5', '4.5', '4', '3.5', '3', '2.5']
df['HSC_m'] = np.select(conditions, choices, default='0')
print(df)
```
